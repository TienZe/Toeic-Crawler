```python
total_chunk = 25

chunk_size = math.ceil(len(collection_infos) / total_chunk)

chunks = [collection_infos[i:i+chunk_size] for i in range(0, collection_infos.shape[0], chunk_size)]
```

```python
processes = []
output_dir = "lesson_chunks"
try:
    for index, chunk in enumerate(chunks):
        output_path = f"{output_dir}/chunk_{index}.csv"
        p = Process(target=get_lessons_of_collection_chunk, args=(chunk, output_path))
        processes.append(p)
        p.start()
        print(f"Started process for chunk {index}")
        sleep(2) # Wait for initializing the driver
        
    for index, p in enumerate(processes):
        p.join()
        print(f"Process {index} finished")
except Exception as e:
    for index, p in enumerate(processes):
        p.terminate()
        print(f"Process {index} finished")
```

```python
for index, p in enumerate(processes):
    p.terminate()
    print(f"Process {index} finished")
```

```python
# combine all output chunks
final_df = pd.DataFrame()

for index, _ in enumerate(chunks):
    output_path = f"{output_dir}/chunk_{index}.csv"
    chunk_df = pd.read_csv(output_path)
    final_df = pd.concat([final_df, chunk_df], ignore_index=True)


final_df.head()
```
